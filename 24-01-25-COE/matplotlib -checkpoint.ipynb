```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y);
plt.title("simple line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='green',linestyle='--',linewidth=2,marker='*') 
plt.title("customized line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.grid(True)  # Show grid
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[200,400,600,800,900,1200]
plt.title("salary growth over 6 months")
plt.plot(x,y,color='green',linestyle='--',linewidth=3,marker='*') 
plt.xlabel("Months")
plt.ylabel("Savings amount")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='skyblue')
plt.title("Basic bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['red','green','blue','yellow','purple'],edgecolor='orange')
plt.title("Basic bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
```

```python
import matplotlib.pyplot as plt
ages=[20,22,22,23,25,27,27,28,29,30,31,32,34,35,35,35,38,40,42,45,45,50,55,65]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
languages=['Ruby','C++','Java','Python']
percentage=[10, 20, 30, 40]
plt.pie(percentage, labels=languages,autopct='%1.1f%%',startangle=90,colors=['pink','green','yellow','orange'])
plt.legend()
plt.title("Programming Language Usage")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot")
plt.xlabel("x vallues")
plt.ylabel("y values")
plt.show()
```
