```python
import matplotlib.pyplot as plt
months=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales=[1200,1500,1700,1600,1800,1900,2000,2100,2300,2500,2400,2600]
plt.plot(months, sales, marker='o', linestyle='--', color='b')
plt.title('Monthly Sales Tracking')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
orderquantities=[5,10,15,5,20,25,10,15,5,30,40,15,20,30,25]
plt.hist(orderquantities,bins=5,color='g',edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Customer ID':[1, 2, 3, 4, 5, 6, 7],
    'Age':[25, 34, 45, 60, 23, 30, 55],
    'Spending':[200, 350, 500, 150, 100, 400, 600]
}
df=pd.DataFrame(data)
plt.figure(figsize=(10,6))
plt.hist(df['Age'],bins=5,color='skyblue',edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Spending'], color='orange')
plt.title('Age and Spending Pattern')
plt.xlabel('Age')
plt.ylabel('Spending')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Student Name': ['yaswitha','ananya','harshitha','sreeja','laxmi'],
    'c++': [85, 70, 95, 60, 88],
    'java': [90, 65, 92, 70, 91],
    'python': [78, 80, 88, 72, 85],
    'mongodb': [88, 75, 89, 65, 90]
}

df=pd.DataFrame(data)
df.set_index('Student Name', inplace=True)
df.plot(kind='bar', figsize=(10, 6), colormap='Set2')
plt.title('Student Marks over Subjects')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.legend(title='Subjects')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
plt.boxplot([df['c++'],df['java'],df['python'],df['mongodb']], labels=['c++','java','python','mongodb'],)
plt.title('Outliers in Marks Across Subjects')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.tight_layout()
plt.show()
```
